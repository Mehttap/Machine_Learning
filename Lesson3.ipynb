The append() method makes it easy to build lists dynamically. For example, you can start with an empty list and then add items to the list using a series of append() statements.

```python
#create an empty list and add items

japanese_fruit_names= []

japanese_fruit_names.append('ringo')
japanese_fruit_names.append('ichigo')
japanese_fruit_names.append('suika')

print(japanese_fruit_names)
```

**Removing Elements from a List**

**Removing an Item Using the del Statement**

If you know the position (index) of the item you want to remove from a list, you can use the del statement.

For Example:

```python
#Removing items in the list

lessons= ['art', 'psychology', 'music', 'dance']
print(lessons)

del lessons [3]
print(lessons)
```

**Removing an Item Using the pop() Methods**

`pop()` metodu listedeki son öğeyi kaldırır, ancak kaldırdıktan sonra o öğeyle çalışmanıza olanak tanır. Yani liste içerisinden çıkarıyor silmiyor.


```python
#Let's pop a lesson from the list of lessons

lessons= ['art', 'psychology', 'music', 'dance']
print(lessons)

popped_lesson= lessons.pop()
print(lessons)
print(popped_lesson)
```

**Popping Items from any Position in a List**

```python
lessons= ['art', 'psychology', 'music', 'dance']

first_taken_lesson= lessons.pop(0)
print('The first lesson I took was an ' + first_taken_lesson.title() + '.')
```

**Removing an Item by Value**

if you only know the value of the item you want to remove, you can use the remove() method.

```python
#Removing item with remove() method

lessons= ['art', 'psychology', 'music', 'dance']
print(lessons)

lessons.remove('music')
print(lessons)
```

**Organizing Lists**

**Sorting a List Permanently with the sort() method**

Liste sıralama methodu


```python
#Alfabetik olarak araba listesini sıralama sort() methodu

cars= ['bwm', 'mercedes', 'audi', 'honda']
cars.sort()  #listeyi alfabetik sırala
print(cars)
```

```python
#Alfabetik sırasının tersine sıralama reverse=True

cars= ['bmw', 'mercedes', 'audi', 'honda']
cars.sort(reverse=True)
print(cars)
```

**Sorting a List Temporarily with the sorted() Function**

```python
#Listeyi geçici sıralama sorted() fonksiyonu

cars= ['bmw', 'mercedes', 'audi', 'honda']
print("Here is the original list:")
print(cars)


print("\nHere is the sorted lists:")
print(sorted(cars))

print("\nHere is the original list again:")
print(cars)
```

**Printing a List in Reverse Order**

reverse() = listeyi tersten yazar

```python
cars= ['bmw', 'audi', 'mercedes', 'honda']
print(cars)

cars.reverse()
print(cars)
```
